# Numpy - Arreglos de datos multidimensionales

Numpy, es el paquete fundamental para computacion cientifica en python. Es una libreria de python que provee un objeto array multidimensional, ademas de varios objetos derivados (como matrices).Tambien provee una serie de rutinas para operaciones sobre arrays.Ademas podemos encontrar modulos para transformada de fourier, algebra lineal basica, operaciones dE estadistica basica, numeros aleatorios y mas.

En el core de Numpy se encuentra el objeto "ndarray". Este encapsula arrays ndimensionales de tipos de datos homogeneos, ademas muchas operaciones hechas en codigo compilado para mejorar la performance.Ahora veamos algunas diferencias importantes entre los arrays de Numpy y las secuencias entandars de python (listas):
- Los arrays de Numpy tienen un tamaño fijo, definido en el momento de su creacion, a diferencia de las listas de python que pueden crecer dinamicamente.
- Los elementos de los arrays de Numpy deben ser del mismo tipo, eso asegura que ocupen el mismo espacio en memoria.
- Los arrays de Numpy proveen varios tipos de operaciones que se ejecutan sobre grandes volumenes de datos, tipicamente esto se realiza de manera mas eficiente y con menos codigo, en comparacion a si se utilizaran secuencias definidas dentro de los built-in de python.

# Criba de Eratostenes

La criba de Eratostenes, es un algoritmo por el cual podemos identificar los numeros primos en un rango entre 2 y n.

El algoritmo es el siguiente:
>Eccribir todos los numeros naturales desde 2 hasta n
>Para i desde 2 hasta sqrt(2) hacer
>>si i no esta marcado hacer
>>>para j desde i hasta n/i hacer
>>>>poner una marca en ixj

```python
from IPython.display import HTML
HTML('<iframe src="eratostenes.html" width=800 height=600></iframe>')
```

```python
import numpy as np
import time
t1 = time.time()
ran = np.arange(100)
es_primo = np.ones((100,),dtype=bool)
es_primo[:2] = 0
maximo = int(np.sqrt(len(es_primo)))
for i in np.arange(2,maximo):
    if (es_primo[i] != False):
        lim = np.arange(i,len(es_primo) / i)
        f = i * lim
        es_primo[f.astype(int)] = False
print(ran[es_primo])
t2 = time.time()
print(t2-t1)
```

# Ejemplo : Difusion usando el algoritmo Random Walk

El algoritmo Random walk unidimensional, que es lo que vamos a realizar consiste en lo siguiente:
* Se tiene un arecta numerica que se extiende en ambas direcciones.
* Comenzamos en el punto "0" de la recta
* Las posibilidades de desplazamiento son dos (derecha o izquierda) ambas tienen la misma probabilidad.

Luego podemos hallar la distancia tipica desde el origen hasta el punto final, luego de "t" pasos.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
numero_recorridos = 1000 # numero de rrecorridos
tiempo_maximo = 200 # tiempo de duracion de cada recorrido
# elegimos al azar los pasos 1 o -1 para el recorrido
tiempo = np.arange(tiempo_maximo)
pasos = 2 * np.random.random_integers(0, 1, size=(numero_recorridos, tiempo_maximo)) - 1
print(np.unique(pasos))# Verificando que todos los pasos son 1 o -1
#<---
# la funcion unique() retorna todos los valores diferentes que contiene un array, si se repite alguno
# solo lo retorna una vez indicando que existe
#--->
# contruimos los recorridos como la suma de los pasos a lo largo del tiempo
posiciones = np.cumsum(pasos, axis=1) # eje = 1: es la dimensión tiempo
#<---
# la funcion cumsum() devuelve la suma acumulada de los elementos de un array
# ejm: a = np.array([1,2,3]) entonces np.cumsum(a) = [1,3,6]
#--->
distancias_al_cuadrado = posiciones**2 # asi lo indica el algoritmo
# obtenemos la media en el eje de los recorridos
media_distancias_al_cuadrado = np.mean(distancias_al_cuadrado, axis=0)
#<---el siguiente codigo corresponde a matplotlib, libreria que aprenderemos a utilizar mas adelante
plt.figure(figsize=(8, 6))
plt.plot(tiempo, np.sqrt(media_distancias_al_cuadrado), 'g.', tiempo, np.sqrt(tiempo), 'y-')
plt.xlabel(r'$tiempo$')
plt.ylabel(r'$\sqrt{\Delta x^{2}}$')
plt.show()
#--->
```

# Broadcasting

- las operraciones basicas se llevan a cabo elemento a elemento
- esto funciona cuando los arrays son del mismo tamaño

**No obstante!**, también es posible hacer operaciones con matrices de diferentes tamaños. Sí Numpy puede transformar estos arreglos en arreglos del mismo tamaño: esta conversión se llama **broadcasting**.

La siguiente imagen muestra un ejemplo de broadcasting

```python
from IPython.display import Image
Image(filename='numpy_broadcasting.png')
```

```python
# comprobemos esto
import numpy as np
a = np.tile(np.arange(0, 40, 10), (3, 1)).T
#<---
# la funcion tile(a,rep) repite "a" tantas veces como lo indica "rep"
# en nuestro caso repite el rango de 0 a 40 cada 10, 3 veces en las filas y una vez en la columna 
#--->
print(a)
b = np.array([0, 1, 2]) # definimos otro array de dimension 1
c = a + b
print("el resultado es : \n",c)
```

un truco util

```python
import numpy as np
ar = np.arange(0,40,10)
print(ar) # imprimimos el array
print("dimension : ",np.shape(ar)) # imprimimos la dimension
#<---
# Podemos agregar una dimension a un array usando la funcion
# np.newaxis()
#--->
br = ar[:,np.newaxis]
print(br)
print("dimension : ",np.shape(br))
b = np.array([0, 1, 2])
print("suma : \n",a + b)
```
